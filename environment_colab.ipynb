### Create a new mamba environment from file `environment.yml`.

File `environment.yml` contains the packages list: [Conda envs documentation](https://mamba.io/docs/using/envs.html).

````python
%%shell
# Ensure the updated PATH is loaded in this shell session
source ~/.bashrc
# Install mamba - if it's not already installed
if ! command -v mamba &> /dev/null
then
    echo "Mamba not found. Installing..."
# Add -y to automatically answer yes to prompts
    conda install -n base -c conda-forge mamba -y
fi

# Ensure conda is initialized
conda init bash

# Reactivate the current shell to load the updated PATH
# source ~/.bashrc

# Add mamba to the PATH in the current shell session
export PATH="$HOME/mambaforge/bin:$PATH"

# Check if mamba is installed and in the path and check conda version
which mamba
conda -V

# Create the environment with more detailed output and specify the solver
# Added the full path to mamba just in case it's not in the PATH
mamba env create -f environment.yml -v --debug --solver=libmamba

# Check environment info
mamba info --envs

# Activate environment ProSeisSN -> not working
# mamba activate ProSeisSN
````
